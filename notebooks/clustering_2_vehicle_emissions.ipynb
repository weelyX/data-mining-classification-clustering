```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


file_path = "../data/vehicle_emissions_clustering.csv"
df = pd.read_csv(file_path)


X = StandardScaler().fit_transform(df.drop(columns=["charges"]))

# ====================================================================
k_values = [2, 3, 4, 5]
costs = []

print("--- Inertia Cost for each K value ---")
for k in k_values:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)

    costs.append(km.inertia_)
    print(f"K = {k} | Cost = {km.inertia_:.2f}")


# =======================================================================
plt.figure(figsize=(8, 5))
plt.plot(k_values, costs, marker="o", linestyle='-', color='blue')

plt.title("Elbow Method For Optimal k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Cost (Inertia)")
plt.grid(True)

plt.show()
```
